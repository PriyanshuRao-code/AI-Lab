```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
train=pd.read_csv('train.csv')
valid=pd.read_csv('valid.csv')
test=pd.read_csv('test.csv')
```

```python
train.columns
```

```python
X_train = train['PT08.S1(CO)'].values.reshape(-1,1)
y_train = train['CO(GT)']
```

```python
X_valid = valid['PT08.S1(CO)'].values.reshape(-1,1)
y_valid = valid['CO(GT)']
```

```python
X_test = test['PT08.S1(CO)'].values.reshape(-1,1)
y_test = test['CO(GT)']
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_train_pred= regressor.predict(X_train)
y_valid_pred= regressor.predict(X_valid)
y_test_pred= regressor.predict(X_test)
```

```python
plt.scatter(X_train, y_train, color="Blue", label="Actual plot")
plt.plot(X_train, y_train_pred, color="Red", label="Predicted plot")
plt.xlabel("Training data: independent variable  (PT08.S1(CO))")
plt.ylabel("Training data: dependent variable (CO(GT))")
plt.title("Training data analysis")
plt.legend()
plt.plot()
```

```python
y_train_rmse = sqrt(mean_squared_error(y_train, y_train_pred))
y_valid_rmse = sqrt(mean_squared_error(y_valid, y_valid_pred))
y_test_rmse = sqrt(mean_squared_error(y_test, y_test_pred))
```

```python
print(f"RMSE for training data: {y_train_rmse}")
print(f"RMSE for validation data: {y_valid_rmse}")
print(f"RMSE for test data: {y_test_rmse}")
```

```python
plt.scatter(y_test, y_test_pred, color="Blue", label="Test data")
plt.xlabel("Test data: original output")
plt.ylabel("Test data: predicted output")
plt.title("Test data analysis")
plt.plot()
```
