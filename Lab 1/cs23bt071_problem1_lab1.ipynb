```python
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('pima-indians-diabetes.csv')
print(df.head(10))
df.info()
df.describe()
print("\n\nMean of the data\n\n",df.mean())
print("\n\nMedian of the data\n\n",df.median())
print("\n\nMode of the data\n\n",df.mode())
print("\n\nMinimum of the data\n\n",df.min())
print("\n\nMaximum of the data\n\n",df.max())
print("\n\nStandard Deviation of the data\n\n",df.std())
print("\n\nQuantile 1\n\n",df.quantile(0.25))
print("\n\nQuantile 3\n\n",df.quantile(0.75))

print("\n\nInterquartile range: \n\n",df.quantile(0.75) - df.quantile(0.25))



# plt.scatter('skin','Age','BMI','pregs','plas')
plt.figure(figsize=(9,5))

plt.subplot(1,2,1)
plt.scatter(df['skin'],df['Age'],marker='o',color='red')
plt.xlabel('Skin')
plt.ylabel('Age')
plt.title('Skin Age relationship')

plt.subplot(1,2,2)
plt.scatter(df['skin'],df['BMI'],marker='o',color='blue')
plt.xlabel('Skin')
plt.ylabel('BMI')
plt.title('Skin BMI relationship')
plt.show()

plt.subplot(1,2,1)
plt.scatter(df['skin'],df['pregs'],marker='o',color='red')
plt.xlabel('Skin')
plt.ylabel('pregs')
plt.title('Skin pregs relationship')

plt.subplot(1,2,2)
plt.scatter(df['skin'],df['plas'],marker='o',color='blue')
plt.xlabel('Skin')
plt.ylabel('plas')
plt.title('Skin plas relationship')
plt.show()


plt.subplot(1,2,1)
plt.scatter(df['Age'],df['BMI'],marker='o',color='red')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Age BMI relationship')

plt.subplot(1,2,2)
plt.scatter(df['Age'],df['pregs'],marker='o',color='blue')
plt.xlabel('Age')
plt.ylabel('pregs')
plt.title('Age pregs relationship')
plt.show()

plt.subplot(1,2,1)
plt.scatter(df['Age'],df['plas'],marker='o',color='red')
plt.xlabel('Age')
plt.ylabel('plas')
plt.title('Age plas relationship')

plt.subplot(1,2,2)
plt.scatter(df['BMI'],df['pregs'],marker='o',color='blue')
plt.xlabel('BMI')
plt.ylabel('pregs')
plt.title('BMI pregs relationship')
plt.show()

plt.subplot(1,2,1)
plt.scatter(df['BMI'],df['plas'],marker='o',color='red')
plt.xlabel('BMI')
plt.ylabel('plas')
plt.title('BMI plas relationship')

plt.subplot(1,2,2)
plt.scatter(df['pregs'],df['plas'],marker='o',color='blue')
plt.xlabel('pregs')
plt.ylabel('plas')
plt.title('pregs plas relationship')
plt.show()


sns.histplot(x='pregs',bins=10,kde=True,data=df)
plt.title("Histplot of pregs")
plt.show()

sns.histplot(x='plas',bins=10,kde=True,data=df)
plt.title("Histplot of plas")
plt.show()

sns.histplot(x='pres',bins=10,kde=True,data=df)
plt.title("Histplot of pres")
plt.show()

sns.histplot(x='skin',bins=10,kde=True,data=df)
plt.title("Histplot of skin")
plt.show()

sns.histplot(x='test',bins=10,kde=True,data=df)
plt.title("Histplot of test")
plt.show()

sns.histplot(x='BMI',bins=10,kde=True,data=df)
plt.title("Histplot of BMI")
plt.show()

sns.histplot(x='pedi',bins=10,kde=True,data=df)
plt.title("Histplot of pedi")
plt.show()

sns.histplot(x='Age',bins=10,kde=True,data=df)
plt.title("Histplot of Age")
plt.show()

sns.histplot(x='class',bins=10,kde=True,data=df)
plt.title("Histplot of class")
plt.show()

grouped=df.groupby('class')
print(grouped.first(10))

from sklearn.datasets import load_iris
iris_group = df.groupby('class')[['BMI', 'Age','plas','pres']].sum()
iris_group.plot(kind="bar")
plt.title("Sum of BMI, Age, plas and pres by 'class' Class")
plt.xlabel("'Class' Class")
plt.ylabel("Sum of Features")
plt.show()


sns.boxplot(x='class', y='pregs',data=df)
plt.title("pregs vs class")
plt.show()

sns.boxplot(x='class', y='plas',data=df)
plt.title("plas vs class")
plt.show()


sns.boxplot(x='class', y='pres',data=df)
plt.title("pres vs class")
plt.show()


sns.boxplot(x='class', y='skin',data=df)
plt.title("skin vs class")
plt.show()


sns.boxplot(x='class', y='test',data=df)
plt.title("test vs class")
plt.show()


sns.boxplot(x='class', y='BMI',data=df)
plt.title("BMI vs class")
plt.show()


sns.boxplot(x='class', y='pedi',data=df)
plt.title("pedi vs class")
plt.show()


sns.boxplot(x='class', y='Age',data=df)
plt.title("Age vs class")
plt.show()
```
