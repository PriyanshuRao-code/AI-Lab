```python
iris = load_iris()

# Convert the features to a DataFrame
data_features = pd.DataFrame(iris['data'], columns=iris['feature_names'])

# Calculate Quartiles
Q1 = data_features.quantile(0.25)
Q2 = data_features.quantile(0.5)
Q3 = data_features.quantile(0.75)

# Calculate Interquartile Range (IQR)
IQR = Q3 - Q1

# Print results
print(f"Q1 (First Quartile):\n{Q1}")
print(f"Q2 (Second Quartile / Median):\n{Q2}")
print(f"Q3 (Third Quartile):\n{Q3}")
print(f"Interquartile Range (IQR):\n{IQR}")
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                    columns= iris['feature_names'] + ['target'])
```

```python

# Group by target and sum the selected features
iris_group = iris.groupby('target')[['petal length (cm)', 'petal width (cm)']].sum()

# Plot the results
iris_group.plot(kind='bar', color=['teal', 'orange', 'red'])

# Show plot
plt.title("Sum of Petal Length and Petal Width by Target Class")
plt.xlabel("Target Class")
plt.ylabel("Sum of Features")
plt.show()
```

```python
sns.pairplot(iris, hue='target', palette = 'dark')
```

```python
sns.histplot(data=iris, x='sepal length (cm)', hue='target', kde=True, palette='coolwarm')
plt.show()
```

```python
sns.set(style="whitegrid")
numerical_columns = iris.columns[:-1]  # Get columns from df2, exclude last

# Loop and plot boxplots for df2 columns
for column in numerical_columns:
    plt.figure(figsize=(10, 6))

    # Access data from df2, not data
    sns.boxplot(x=iris[column], color='skyblue')

    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.ylabel('Value')
    plt.tight_layout()
    plt.show()
```
